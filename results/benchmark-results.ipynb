```python
from wandb import Api

api = Api()
```

```python
runs = api.runs("martenlienen/torchode", filters={"group": "cnf"})
```

```python
rows = []
for r in runs:
    df = r.history(keys=["epoch", "val/loss", "val/fw_steps", "val/fw_time", "val/fw_time_per_step", "val/fw_model_time_per_step", "val/fw_loop_time"])
    if len(df) == 0:
        continue
    df = df.set_index("epoch")
    row = df.loc[r.summary["best_epoch"]].copy()
    row["library"] = r.config["library"]
    rows.append(row)
df = pd.DataFrame(rows)
```

```python
df.groupby("library").mean()
```

```python
df.groupby("library").std()
```
